```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://quality.data.gov.tw/dq_download_csv.php?nid=6353&md5_url=e14267c580987dfa9410e4cb91166098")
```

```python
df.head()
```

## 取得各縣市總和

```python
df_county = df.groupby('County').sum()
```

```python
df_county
```

## 為縣市做排名

```python
df_county_R = df_county.sort_values(by = ['SampleValue'], ascending = False)
```

```python
df_county_R
```

###發現西南部數值較高
